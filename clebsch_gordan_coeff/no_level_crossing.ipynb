```python
def_list = np.linspace(0,1,100)
def gen_mat(x, param):
    return [[1+x, param], [param, 2-x]]
```

```python
mats_list = [gen_mat(x, 0.1) for x in def_list]
spec = np.sort(np.array([np.linalg.eigvals(mat) for mat in mats_list]))
plt.plot(spec[:,0])
plt.plot(spec[:,1])
```
